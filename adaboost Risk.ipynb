```python
start = time.time()
# params = {'n_estimators':[5,15],'max_depth':[100,200],'min_samples_leaf':[30,100],'loss':['exponential','deviance']}
params = {'n_estimators':[30,50],'max_depth':[80,130,200],'min_samples_leaf':[10,30,50]}
gbc = GradientBoostingClassifier(warm_start=False)
print('using warm_start = FALSE')
clf = GridSearchCV(gbc,params) #cv=None defaults to 5, 
clf.fit(x[feat],x['winner'])
min = (time.time() - start) / 60
print(f'it took {round(min,2)} minutes to do the grid search')
```

# train on all data that ends before 400 turns, then predict over time for all games that go between 400-500 turns long.

This means train on 15k games, and test on 2505

```python
#turn_lower_bound,turn_upper_bound specified in several cells above
time_sample =60
results = dict()
# list of acceptable turns
summary = pd.read_csv('../../402_project/data/big2_summary.csv')
a = summary['num']
turns_to_consider = a[np.linspace(1,len(a),time_sample).astype(int) - 1]
print(turns_to_consider)

for i,t in enumerate(turns_to_consider):
    try:
        df = pd.read_hdf('../../402_project/data/01_big_nth_turn_datasets/big2/' + f'big_{t}_turn.hdf')

        num_games = df.shape[0]
        stalemates = len(df[df.winner.isna()].index)
        df.drop(index=df[df.winner.isna()].index,inplace=True)
        df = df.loc[(df.total_numer_turns_in_game > turn_lower_bound) & (df.total_numer_turns_in_game < turn_upper_bound)].copy()
        #adaboost 
        score = clf.score(df[feat],df.winner)
        results[i] = {'turn':t
                     ,'score':score
                     ,'game_count_without_stalemates':num_games - stalemates
                     ,'stalemate':stalemates
                     ,'num_games':num_games
                     }

        print('done with turn',t)
    except:
        pass
    
month, day = dt.datetime.now().month, dt.datetime.now().day
hour, minute = dt.datetime.now().hour, dt.datetime.now().minute
title = f'risk_adaboost_summary_{month}_{day}_{hour}_{minute}'
keep = pd.DataFrame(results).T
keep.to_csv('../../402_project/data/'+ title + '.csv',index=False)
```
